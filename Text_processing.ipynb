### Remove URL's

```python
def remove_url(texts): 
    pattern=re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'',texts)
```

### 3rd step is Remove Punctuation

```python
import string,time
string.punctuation
```

```python
exclude=string.punctuation
```

```python
def remove_punc(text):
    for char in exclude:
        text = text.replace(char, ' ')
    return text
```

## Chat word treatment

```python
slang_dict = {}

with open("slang.txt", "r", encoding="utf-8") as file:
    for line in file:
        line = line.strip()
        if not line or "=" not in line:
            continue
        key, value = line.split("=", 1)
        slang_dict[key.strip()] = value.strip()
```

```python
slang_dict
```

```python
def chat_conversion(text):
    new_text = []
    
    for word in text.split():
        if word.upper() in slang_dict:
            new_text.append(slang_dict[word.upper()])
        else:
            new_text.append(word)
    
    return " ".join(new_text)
```

```python
chat_conversion('WTF you are doing')
```

```python
chat_conversion('Plese W8 for me ')
```

```python
 from nltk.corpus import stopwords
def remove_stopwords(text):        
    stop_words = set(stopwords.words('english')) # Convert to set for faster lookup
    words = [word for word in text.split() # List comprehension for cleaner code
             if word.lower() not in stop_words]
    return " ".join(words)


# Alternet 
    
    # def remove_stopwords(text):
    # from nltk.corpus import stopwords
    # stop_words = set(stopwords.words('english'))
    
    # new_text = []
    # for word in text.split():
    #     if word.lower() not in stop_words:  # Convert to lowercase for matching
    #         new_text.append(word)
    
    # return " ".join(new_text)
```

## Handling emojies

```python
#  To remove emoji 
import emoji

def remove_emoji(text):
    return emoji.replace_emoji(text, replace=' ') # Remove all emojis

def remove_emoji_simple(text):# Or if you want to keep only text
    return emoji.demojize(text)
```

```python
text_with_emojis = "Hello! ðŸ˜Š I'm feeling great today! ðŸŽ‰ Let's meet at 5pm! ðŸ“…"
print("Original:", text_with_emojis)
print("Cleaned:", remove_emoji(text_with_emojis))
print("Cleaned:", remove_emoji_simple(text_with_emojis))

```

## Stemming 

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
def stem_words(text): 
    return " ".join([ps.stem(word)for word in text.split()])
```

```python
sample= "Walk walks walking walked"
stem_words(sample)
```
