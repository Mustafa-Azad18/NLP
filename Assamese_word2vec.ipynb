```python
import numpy as np 
import pandas as pd 
import gensim 
import os 
```

```python
import os
import re

story = []
folder = "Assamese"

def assamese_tokenize(text):

    tokens = re.findall(r'[\u0980-\u09FF]+', text)
    return tokens

for filename in os.listdir(folder):
    filepath = os.path.join(folder, filename)
    if os.path.isfile(filepath):
        with open(filepath, encoding="utf-8", errors="ignore") as f:
            corpus = f.read()
        lines = corpus.splitlines()
        for line in lines:
            tokens = assamese_tokenize(line)
            if tokens:
                story.append(tokens)
```

```python
from gensim.models import Word2Vec

model = Word2Vec(
    sentences=story,
    vector_size=100,
    window=5,
    min_count=2,
    workers=4,
    sg=1,
    epochs=100,        
    hs=0,             # Negative sampling (faster, better for large corpus)
    negative=10,      # 10 negative samples per positive (good for small corpus)
    alpha=0.025,      # Initial learning rate
    min_alpha=0.0001  # Learning rate decay
)

# Save the model
model.save("assamese_word2vec.model")
print(f"Vocabulary size: {len(model.wv)}")
```

```python
model.wv.most_similar('মহিলা')
```

```python

word = 'ভাষা'  
if word in model.wv:
    similar = model.wv.most_similar(word, topn=10)
    for w, score in similar:
        print(f"{w}: {score:.4f}")
else:
    print(f"'{word}' not in vocabulary")

# Check vocabulary
print(list(model.wv.index_to_key[:20]))  

# Get vector for a word
vector = model.wv['ভাষা']
print(vector.shape)  # Should be (100,)
```
